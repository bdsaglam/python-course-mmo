## A tale of two APIs - MATLAB vs Object Oriented

```python
X = np.linspace(- np.pi, np.pi, 256 , endpoint = True)
C, S = np.cos(X), np.sin(X)
```

### MATLAB API

```python
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

```python
# Create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize = ( 12, 6 ), dpi = 80)

# Create a new subplot from a grid of 1x1
plt.subplot(1 , 1 , 1 )

# Plot cosine with a blue continuous line of width 1 (pixels)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine" )

# Set x limits
plt.xlim(-4.0, 4.0)

# Set y limits
plt.ylim(-1.0, 1.0 )


plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.yticks([-1, 0, +1], 
           [r'$-1$', r'$0$' , r'$+1$' ])

plt.legend(loc ='upper left')

plt.show()
```

### Object Oriented API

```python
fig, ax = plt.subplots(figsize=(12,6), dpi=80)

# Plot cosine with a blue continuous line of width 1 (pixels)
ax.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
ax.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine" )
ax.legend(loc ='upper right')

# Set x and y axes limits
ax.set(xlim=(-3.15, 3.15), ylim=(-1.0, 1.0))

ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

ax.set_yticks([-1, 0, +1])
ax.set_yticklabels([r'$-1$', r'$0$' , r'$+1$' ]);
```

```python
fig, ax = plt.subplots(figsize=(12,6), dpi=80)

# Plot cosine with a blue continuous line of width 1 (pixels)
ax.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
ax.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine" )
ax.legend(loc ='upper right')

# Set x and y axes limits
ax.set(xlim=(-3.15, 3.15), ylim=(-1.0, 1.0))

ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

ax.set_yticks([-1, 0, +1])
ax.set_yticklabels([r'$-1$', r'$0$' , r'$+1$' ]);

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom']. set_position(('data' , 0 ))
ax.spines['left'].set_position(('data', 0))
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left' )
```

```python
fig, ax = plt.subplots(figsize=(12,6), dpi=80)

# Plot cosine with a blue continuous line of width 1 (pixels)
ax.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
ax.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine" )
ax.legend(loc ='upper right')

# Set x and y axes limits
ax.set(xlim=(-3.15, 3.15), ylim=(-1.0, 1.0))

ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

ax.set_yticks([-1, 0, +1])
ax.set_yticklabels([r'$-1$', r'$0$' , r'$+1$' ]);

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom']. set_position(('data' , 0 ))
ax.spines['left'].set_position(('data', 0))
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left' )

# Create a figure of size 8x6 inches, 80 dots per inch
theta = 2 * np.pi / 3

ax.plot([theta, theta], [0, np.cos(theta)], color='blue', linewidth=2.5, linestyle="--")
ax.scatter([theta, ], [np.cos(theta), ], 50, color='blue')
ax.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy= (theta, np.sin(theta)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2" ))

ax.plot([theta, theta],[0, np.sin(theta)], color ='red', linewidth=2.5, linestyle="--" )
ax.scatter([theta, ],[np.sin(theta), ], 50, color='red')
ax.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy= (theta, np.cos(theta)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
            )
```

```python
# sin(2*pi*f*t)
t = np.arange(0, 1, 0.002)
signal1 = np.sin(2 * np.pi * 1 * t)
signal2 = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(t.size)
```

```python
# MATLAB API
plt.plot(t, signal1, '-r', t, signal2, '-g')
```

```python
# OO API
fig, ax = plt.subplots()
ax.plot(t, signal1, color='red', linewidth=5, linestyle='--')
ax.plot(t, signal2, color='green')
```

#### Logarithmic Axes

```python
x = np.arange(0.01, 20, 0.01)

fig = plt.figure(figsize=(12,18))

ax1 = fig.add_subplot(311)
y1 = np.exp(x/6.)
ax1.semilogy(x, y1);
ax1.grid(True)
ax1.set_ylabel('Only Y');

ax2 = fig.add_subplot(312)
y2 = np.cos(np.pi*x)
ax2.semilogx(x, y2);
ax2.set_xlim([0.01, 20]);
ax2.grid(True)
ax2.set_ylabel('Only X');

ax3 = fig.add_subplot(313)
y3 = np.exp(x/4.)
ax3.loglog(x, y3, basex=3);
ax3.grid(True)
ax3.set_ylabel('Both X and Y');
plt.show()
```

#### Multiple Y axes

```python
x = np.arange(0.001, np.e, 0.01)
y1 = 10*np.exp(-x)
y2 = np.log(x)

fig = plt.figure(figsize=(12,6))

ax1 = fig.add_subplot(111)
ln1 = ax1.plot(x, y1, label='10exp(x)', color='red', linewidth=2);
ax1.set_ylabel('Y values for 10*exp(-x)');

ax2 = ax1.twinx() # this is the important function
ln2 = ax2.plot(x, y2, label='log(x)', color='blue', linewidth=3);
ax2.set_ylabel('Y values for ln(x)');
ax2.set_xlabel('Same X for both');

lns = ln1+ln2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc='center right')
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)

plt.axes([0.025,0.025,0.95,0.95])
plt.scatter(X,Y, s=75, c=T, alpha=.5, cmap='jet')

plt.xlim(-1.5,1.5), plt.xticks([])
plt.ylim(-1.5,1.5), plt.yticks([])
plt.show()
```

### Bar

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / n) * np.random.uniform(0.5, 1.0, n)
Y2 = -(1 - X / n) * np.random.uniform(0.5, 1.0, n)

fig, ax = plt.subplots(figsize=(12, 6), dpi=80)
ax.bar(X, Y1, facecolor='#9999ff', edgecolor='black')
ax.bar(X, Y2, facecolor='#ff9999', edgecolor='white')

for x, y, y2 in zip (X, Y1, Y2):
    plt.text(x, y + 0.05, '%.2f'%y, ha='center', va='bottom')
    plt.text(x, y2 - 0.15, '%.2f'%y2, ha='center', va='bottom')
    
plt.ylim( -1.25 , + 1.25)
```

### Contours Images

```python
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.axes([0.025,0.025,0.95,0.95])

plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap=plt.cm.hot)
C = plt.contour(X, Y, f(X,Y), 8, colors='black')
plt.clabel(C, inline=1, fontsize=10)

plt.xticks([]), plt.yticks([])
# savefig('../figures/contour_ex.png',dpi=48)
plt.show()
```

```python
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 10
x = np.linspace(-3, 3, 3.5*n)
y = np.linspace(-3, 3, 3.0*n)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)

plt.axes([0.025,0.025,0.95,0.95])
plt.imshow(Z,interpolation='nearest', cmap='bone', origin='lower')
plt.colorbar(shrink=.8)

plt.xticks([]), plt.yticks([])
plt.show()
```

### Polar graphs

```python
ax = plt.subplot(polar=True)

N = 20
theta = np.arange(0.0, 2*np.pi, 2*np.pi/N)
radii = 10*np.random.rand(N)
width = np.pi/4*np.random.rand(N)
bars = plt.bar(theta, radii, width=width, bottom=0.0)

for r,bar in zip(radii, bars):
    bar.set_facecolor( plt.cm.jet(r/10.))
    bar.set_alpha(0.5)

ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()
```

### 3D graphs

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.jet)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.jet)
ax.set_zlim(-2,2)

# savefig('../figures/plot3d_ex.png',dpi=48)
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
```

```python
def currency(x, pos):
    'The two args are the value and tick position'
    if x >= 1000000:
        return '${:1.1f}M'.format(x*1e-6)
    return '${:1.0f}K'.format(x*1e-3)
```

```python
fig, ax = plt.subplots()
top_10.plot(kind='barh', y="Sales", x="Name", ax=ax)
ax.set_xlim([-10000, 140000])
ax.set(title='2014 Revenue', xlabel='Total Revenue', ylabel='Customer')
ax.xaxis.set_major_formatter(FuncFormatter(currency))
ax.legend().set_visible(False)
```

```python
# Create the figure and the axes
fig, ax = plt.subplots()

# Plot the data and get the averaged
top_10.plot(kind='barh', y="Sales", x="Name", ax=ax)
avg = top_10['Sales'].mean()

# Set limits and labels
ax.set_xlim([-10000, 140000])
ax.set(title='2014 Revenue', xlabel='Total Revenue', ylabel='Customer')

# Add a line for the average
ax.axvline(x=avg, color='b', label='Average', linestyle='--', linewidth=1)

# Annotate the new customers
for cust in [3, 5, 8]:
    ax.text(140000, cust, "New Customer")

# Format the currency
ax.xaxis.set_major_formatter(FuncFormatter(currency))

# Hide the legend
ax.legend().set_visible(False)
```

```python
# Get the figure and the axes
fig, (ax0, ax1) = plt.subplots(nrows=1,ncols=2, sharey=True, figsize=(12, 6))

# plot sales
top_10.plot(kind='barh', y="Sales", x="Name", ax=ax0)
ax0.set_xlim([-10000, 140000])
ax0.set(title='Revenue', xlabel='Total Revenue', ylabel='Customers')

# Plot the average as a vertical line
avg = top_10['Sales'].mean()
ax0.axvline(x=avg, color='b', label='Average', linestyle='--', linewidth=1)

# Format the currency
ax0.xaxis.set_major_formatter(FuncFormatter(currency))

# plot purchases
top_10.plot(kind='barh', y="Purchases", x="Name", ax=ax1)
avg = top_10['Purchases'].mean()
ax1.set(title='Units', xlabel='Total Units', ylabel='')
ax1.axvline(x=avg, color='b', label='Average', linestyle='--', linewidth=1)

# Title the figure
fig.suptitle('2014 Sales Analysis', fontsize=14, fontweight='bold');

# Hide the legends
ax1.legend().set_visible(False)
ax0.legend().set_visible(False)
```
